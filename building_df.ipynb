```python
names = pd.read_csv("src/babynames.csv")
```

```python
annualnamecounts = names.groupby('Year').sum().reset_index()
```

```python
names['TotalAnnualCounts'] = [annualnamecounts[annualnamecounts['Year']==y] for y in names['Year']]
```

```python
names['AnnualFreq']=names['Count']/annualnamecounts[annualnamecounts['Year']==names['Year']]['Count']
```
